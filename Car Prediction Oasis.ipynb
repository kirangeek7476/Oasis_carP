```python
# Bivariate Analysis
def analysis(data_frame,x_axis_feature,y_axis_feature,k):
    fig, axs = plt.subplots(1, 3, figsize=(20,5) )
    axs = axs.flatten()
    sns.scatterplot(x=col[k],   y=y_axis_feature, hue="price", data=data_frame, ax=axs[0])
    sns.scatterplot(x=col[k+1], y=y_axis_feature, hue="price", data=data_frame, ax=axs[1])
    sns.scatterplot(x=col[k+2], y=y_axis_feature, hue="price", data=data_frame, ax=axs[2])
    #axs[3].set_visible(False)
    
    plt.show()


analysis(df,col,"price",0)
    
```

# Correlated Features

```python
#Correlation with output variable
cor_target = correlation['price']
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.4]
relevant_features
```

```python
df_main_features = df[['wheelbase','carlength','carwidth','curbweight','enginesize','boreratio','horsepower','price']]
df_main_features.head()
```

```python
plt.figure(figsize=(10,10))
correlation = df_main_features.corr(method='pearson')
sns.heatmap(correlation,annot=True,cmap = 'Blues')
```

# Preparing our data for Modelling

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_main_features.drop(['price'],axis=1)
Y = df_main_features['price']
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3)
```

```python
X_train.shape,Y_train.shape
```

```python
X_test.shape,Y_test.shape
```

# Modelling - Linear Regression

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,Y_test)
```

```python
print(f"Accuracy : {model.score(X_test,Y_test)*100:.2f}%")
```

```python
fig = plt.figure()
plt.scatter(Y_test,y_pred)
fig.suptitle('Y_test vs y_pred', fontsize=20)              # Plot heading 
plt.xlabel('Y_test', fontsize=18)                          # X-label
plt.ylabel('Y_pred', fontsize=16) 
```
